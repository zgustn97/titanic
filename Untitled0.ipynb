```python
df_null_count = null_count.reset_index()

df_null_count.head()
df_null_count.columns = ["컬럼명","결측치 수"]
df_null_count.sort_values(by="결측치 수",ascending= False)
df = df.drop(["Ticket"] , axis=1)
# df["Sex"].unique()
df['Sex'] = df["Sex"].map({'male': 1, 'female': 2},int) 
# null_count
df[['Name','Age']]
df['test'] = df['Name'].str.extract('([A-Za-z]+)/.', expand = False)
# df[['test']]
df['test'] = df['Name']
df['test'] = df['test'].str.extract('([A-Za-z]+)\.')
df['test'].value_counts()
Name_mapping = ({'Mr': 1, 'Miss': 2, 'Mrs':1 , 'Master':3, 'Dr': 3 , 'Rev': 3, 'Mile':3 , 'Major':3, 'Col':3, 'Capt':3 , 'Lady':3, 'Johnheer':3 , \
              'Ms':2, 'Mme':3 , 'Countess':3 , 'Don':3 , 'Sir': 3})
df['test'] = df['test'].map(Name_mapping,int)
# df.dtype()
df.loc[ df['Fare'] <=17, 'Fare'] = 0
df.loc[(df['Fare'] > 17) & (df['Fare'] <=30), 'Fare'] = 1
df.loc[(df['Fare'] > 30) & (df['Fare'] <=100), 'Fare'] = 2
df.loc[ df['Fare'] > 100, 'Fare'] = 3


age_mapping = {1: 1 ,2: 2, 3:3}
df.groupby("test")["Age"].mean()
df["Age"] = df["Age"].fillna(df["Age"].mean()).to_frame()

df.loc[ df['Age'] <=17, 'Age'] = 0
df.loc[(df['Age'] > 17) & (df['Age'] <=30), 'Age'] = 1
df.loc[(df['Age'] > 30) & (df['Age'] <=100), 'Age'] = 2
df.loc[ df['Age'] > 100, 'Age'] = 3


df.drop(columns = ["Name"], axis=1)
df["Fare"] = df["Fare"].fillna(df["Fare"].mean()).to_frame()
x = df.drop(columns = ["Survived","Name","Cabin","Embarked","PassengerId","test"])
print(x)
print('\n')
x = torch.FloatTensor(x.values)

y = torch.FloatTensor(df["Survived"].values)


testdf['Sex'] = testdf["Sex"].map({'male': 1, 'female': 2},int) 
testdf["Age"] = testdf["Age"].fillna(testdf["Age"].mean()).to_frame()
testdf["Fare"] = testdf["Fare"].fillna(testdf["Fare"].mean()).to_frame()

testdf.loc[ testdf['Fare'] <=17, 'Fare'] = 0
testdf.loc[(testdf['Fare'] > 17) & (testdf['Fare'] <=30), 'Fare'] = 1
testdf.loc[(testdf['Fare'] > 30) & (testdf['Fare'] <=100), 'Fare'] = 2
testdf.loc[ testdf['Fare'] > 100, 'Fare'] = 3

testdf.loc[ testdf['Age'] <=17, 'Age'] = 0
testdf.loc[(testdf['Age'] > 17) & (testdf['Age'] <=30), 'Age'] = 1
testdf.loc[(testdf['Age'] > 30) & (testdf['Age'] <=100), 'Age'] = 2
testdf.loc[ testdf['Age'] > 100, 'Age'] = 3

dfx = testdf.drop(columns = ["Name","Cabin","Embarked","PassengerId","Ticket"])
x.shape

# dfy = torch.FloatTensor(dfx["Survived"].values)
dfx = torch.FloatTensor(dfx.values)


Linear = torch.nn.Linear(6,1)
ac = torch.nn.Sigmoid()




optim = torch.optim.SGD(Linear.parameters() ,lr = 0.001)



y = y.reshape(891,1)
 
 
loss_f = torch.nn.BCELoss()



for iter in range (40):
  for i in range (1000):
    optim.zero_grad()
    forward = ac(Linear(x))
    loss = loss_f(forward , y)
    
    loss.backward()
    optim.step()

    # final_train_acc = train_acc/891
    if i%100 == 0:
      print("i: ",i, "loss: ",loss, "wei:",Linear.weight,"b:",Linear.bias ,"acc:",(torch.sum((forward > 0.5).long() == y)/891)*100)

bc = (ac(Linear(dfx))> 0.5).long().detach().numpy()
bc
dfy["Survived"] = bc
dfy = dfy.set_index('PassengerId')
dfy.to_csv('drive/MyDrive/data_set/first_submission.csv')
# print(x)
# print(Linear(x))
# print(ac(Linear(x)))    
# print( (forward > 0.5).long() )
# print(ac(Linear(dfx)))
# print("acc:",(torch.sum((ac(Linear(dfx)) > 0.5).long() == dfy )/418)*100)


```
